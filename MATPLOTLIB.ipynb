```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#grafikleri sürekli ply.show() yazarak ekrana göstermek yerine Jupyter Notebook a
#özel olarak aşağıdaki komutu çalıştırırsak sürekli show() metodunu çalıştırmamıza gerek kalmaz
#hücreyi her çalıştırdığımızda grafikte gösterilir
%matplotlib inline
```

```python
#şimdi numpy arrayı türünde iki array tanımlıyalım
```

```python
#array oluşturmadan önce listeleri oluşturalım 
#sonra bu listeleri array e dönüştürelim
```

```python
ageList=[10,20,30,30,30,40,50,60,70,75]
weightList=[20,60,80,85,86,87,70,90,95,90]
```

```python
#yukarıda tanımladığımız listeleri şimdi arraye çevirelim
ageArray=np.array(ageList)
weightArray=np.array(weightList)
```

```python
ageArray
```

```python
#şimdi bu verileri matplotlib ile grafiğe dökelim
plt.plot(ageArray,weightArray,"r")
#x ekseni=yaş
#y ekseni=kilo
#yaşa göre kiloları gösteren bir grafik
```

```python
#değişken yerleri x ve y eksenlerini ifade eder
#biraz saçma ama eğer kilolara göre yaşı gösteren bir grafik çizmek istersek:
plt.plot(weightArray,ageArray,"g")
```

```python
#eğer x ve y eksenlerini adlandırmak istersek:
plt.plot(ageArray,weightArray,"r")
plt.xlabel("Age")
plt.ylabel("Weight")
#plt.show()
plt.title("Kilonun Yaşa Göre Dağılımı")
```

```python
#üstteki mavi grafik alttaki yeşil grafiği tamamen kapattı aynı grafik üzerinde
#birden fazla sonuç göstermek için ne yapmalıyız????
```

```python
type(plt.subplots())#bir tuple değer döndürüyor def nrows=1,ncols=1n dir
```

```python
(myFigure,myAxes)=plt.subplots(nrows=1,ncols=2)
type(myAxes)#tyoe-->ndarray
type(myFigure)#type-->matplotlib.figure.Figure dir

#burada myAxes bir array olduğundan for döngüsüne sokabiliriz
#for döngüsü içinde eksenler için atamalar yapabiliriz(1 tane x,2 tane y ekseni)
for axes in myAxes:
    axes.plot(numpyArray1,numpyArray2,"g")#grafik çizimi
    axes.set_xlabel("X Ekseni")
    axes.set_ylabel("Y Ekseni")
    
#iki kolon çok içiçe bunu önlemek için:
plt.tight_layout()#for içinde yazınca da aynı sonucu veriyor demekki her döndüğünde aynı işlemi yapıyor 
    
```

GÖRSELLEŞTİRME

```python
#figür üzerindeki işlemlerin hepsi tek hücrede gerçekleştirilmeli 
#çünkü hücre her çalıştırıldığında nesnenin adresi değişiyor

newFigure=plt.figure()#figür nesnesi oluşturuldu

#şimdi bu nesne üzerinden eksenleri oluşturalım
newFigureAxes=newFigure.add_axes([0.1,0.1,0.9,0.9])

#şimdi grafiği oluşturalım
newFigureAxes.plot(numpyArray1,numpyArray1**2,label="numpyArray**2")#bu labellara legend denir
newFigureAxes.plot(numpyArray1,numpyArray1**3,label="numpyArray**3")
newFigureAxes.legend()
```

```python
#eğer bu grafiği kaydetmek istersek:
newFigure.savefig("myFigure.png",dpi=200)#bunu jupyter notebook ile yanı dizine kaydetmiştir
#.png ile resim dosyası olduğunu belirtiriz
#dpi ile de çözünürlüğü ayarlarız
```

COLOR CHANGED

```python
(myFigure,myAxes)=plt.subplots()#parantez içinde de axes boyutları ve başlanıçları belirlenebilir
myAxes.plot(numpyArray1,numpyArray1**2,color="#3A95A8")
myAxes.plot(numpyArray1**2,numpyArray1,color="#C96F23")
```

```python
#çizgi kalınlığını ve stilini de ayarlayabiliriz
```

```python
(newFigure,newAxes)=plt.subplots()
newAxes.plot(numpyArray1,numpyArray1+2,color="Blue",linewidth=1.0)
newAxes.plot(numpyArray1,numpyArray1+3,color="green",linewidth=1.0)

newAxes.plot(numpyArray1,numpyArray1+4,color="#C96F23",linewidth=1.0,linestyle="-.")
newAxes.plot(numpyArray1,numpyArray1+5,color="#C96F23",linewidth=1.0,linestyle=":")
newAxes.plot(numpyArray1,numpyArray1+6,color="#C96F23",linewidth=1.0,linestyle="--")#tek çizgi düz çizgi olarak kabul edilir


newAxes.plot(numpyArray1,numpyArray1+7,color="#000000",linewidth=1.0,linestyle="--",marker="o",markersize=8,markerfacecolor="red")
newAxes.plot(numpyArray1,numpyArray1+8,color="#000000",linewidth=1.0,linestyle="--",marker="+",markersize=4,markerfacecolor="red")
#önceki kodlarda *- şeklinde bir gösterim vardı buradaki yıldızlar marker dır
#markerfacecolor ,markerın içini doldurur
#markersize ile marker büyüklüğü ayarlanır
```
